```python
import numpy as np
import matplotlib.pyplot as plt

#  функция, решающая функцию
def f(x): 
    return np.log10(x) - np.cos(x)

#  параметры метода(интервал с корнем, точность и счётчик итераций)
a, b = 1.3, 1.5
eps = 0.00001
iter_count = 0

#  решение методом бисекции
while b - a > eps:
    c = (a + b) / 2
    if f(a) * f(c) < 0:
        b = c
    else:
        a = c
    iter_count += 1

#  визуализация для корня
x = np.linspace(0.1, 3, 200)
plt.plot(x, f(x), 'b--')
plt.axhline(0, c='k', lw=0.5)
plt.title('Локализация корня')
plt.grid()
plt.plot(c, 0, 'ro', markersize=8, label=f'Корень x = {c:.4f}')
plt.show()

#  вывод результатов
print(f'Результат: x = {c:.4f}')
print(f'Число итераций: {iter_count}')
print(f'Проверка: f({c:.4f}) = {f(c):.6f}')
```

Упражнение 2

```python
import numpy as np
import matplotlib.pyplot as plt

# функция для нахождения уравнения
def func(x):
    return x**2 - 3.2*x - 1

# итерационная функция для первого корня (около -0.3)
def phi1(x):
    return (x**2 - 1) / 3.2

# итерационная функция для второго корня (около 3.5)
# выражаем x = (x² - 1) / 3.2 не подходит, используем x = sqrt(3.2x + 1)
def phi2(x):
    return np.sqrt(3.2*x + 1)

# параметры для первого корня
x0_1 = -0.5
eps = 1e-4
diff = 1
iter_count_1 = 0

# итерации для первого корня
while diff > eps:
    x1_1 = phi1(x0_1)
    diff = abs(x1_1 - x0_1)
    x0_1 = x1_1
    iter_count_1 += 1

# параметры для второго корня
x0_2 = 3.5
diff = 1
iter_count_2 = 0

# итерации для второго корня
while diff > eps:
    x1_2 = phi2(x0_2)
    diff = abs(x1_2 - x0_2)
    x0_2 = x1_2
    iter_count_2 += 1

# визуализация
x = np.linspace(-2, 5, 300)
plt.plot(x, func(x), 'b-', label='f(x)')
plt.plot(x, x*0, 'r--', label='y=0')
plt.grid(alpha=0.3)
plt.title('Локализация корней')

# отметка обоих корней
plt.plot(x1_1, 0, 'ro', markersize=8, label=f'Корень 1: x = {x1_1:.4f}')
plt.plot(x1_2, 0, 'go', markersize=8, label=f'Корень 2: x = {x1_2:.4f}')

plt.legend()
plt.show()

# вывод результатов
print(f'Первый корень: {x1_1:.4f}')
print(f'Итераций для первого корня: {iter_count_1}')
print(f'Второй корень: {x1_2:.4f}')
print(f'Итераций для второго корня: {iter_count_2}')
```

```python
#  определение функций
def f(x):
    return 4*x - np.log(x) - 4 if x > 0 else np.nan
def phi(x):
    return (4 + np.log(x)) / 4

#  решение методом простых итераций
x = 2.0
while True:
    x_new = phi(x)
    if abs(x_new - x) < 1e-4:
        break
    x = x_new

#  визуализация
x_vals = np.linspace(0.1, 3, 300)
plt.plot(x_vals, [f(v) for v in x_vals], 'b-')
plt.axhline(0, c='r', ls='--')
plt.plot(x_new, 0, 'ro', label=f'x = {x_new:.4f}')
plt.grid(alpha=0.3)
plt.title('4x - ln x = 4')
plt.legend()
plt.show()

#  вывод
print(f'Корень: {x_new:.4f}')
```
