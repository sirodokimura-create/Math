```python
#  1.
from numpy import *
from scipy.optimize import fsolve

def n(x):
    return log(x + 5) - cos(x)

root = fsolve(n, -4.35)
print('Корень (округлённый):', root[0])
```

```python
#  2 б)
from math import *
def otvet(a, b, eps):
    while b - a > eps:
        c = (a + b) / 2
        if log(c + 5) - cos(c) == 0:
            return c
        elif log((a + 5) - cos(a)) * (log(c + 5) - cos(c)) < 0:
            b = c
        else:
            a = c
    return (a + b) / 2
print(otvet(-4.45, -4.3, 1e-3))
```

```python
#  3
def iteration(x0, eps=1e-6, max_iter=100):
    x = x0
    for i in range(max_iter):
        x_new = -5 + exp(cos(x))  
        if abs(x_new - x) < eps:
            return x_new, i + 1
        x = x_new
    return x, max_iter

x0 = -4.35
root, n = iteration(x0)
print(f'Корень: {root:.8f}, итераций: {n}')
```
