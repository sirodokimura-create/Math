Упражнение 1

```python
import numpy as np
import matplotlib.pyplot as plt

print("Решение уравнения 0,5*cos(x) + 1 = 0")

#  функция, решающая уравнение
def f(x): 
    return 0.5 * np.cos(x) + 1

#  построение графика
x = np.linspace(0, 2, 100)
plt.plot(x, f(x), 'g-', label='φ(x)')
plt.plot(x, x, 'r--', label='y=x')
plt.grid(True, alpha=0.3)

#  начальное приближение
x0 = 0.5
x_points = [x0]

#  визуализация итераций
for i in range(6):
    x_next = f(x_points[-1])
    plt.plot([x_points[i], x_points[i]], [x_points[i], x_next], 'g-')
    plt.plot([x_points[i], x_next], [x_next, x_next], 'g-')
    plt.plot(x_points[i], f(x_points[i]), 'ro', markersize=6)
    x_points.append(x_next)

#  оформление графика
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.axis([0, 2, 0, 2])
plt.gca().set_aspect('equal')
plt.show()

#  вывод
print("Итерации:", [f"{i}: {val:.6f}" for i, val in enumerate(x_points)]) 
```

Упражнение 2

```python
from tabulate import tabulate

print("Решение уравнения x*sin(x) - 1 = 0")
print("Метод простой итерации, точность 10**(-5)")

#  функция решения уравнения
def phi(x):
    return x - 0.5 * (x * np.sin(x) - 1)

#  начальное приближение
x = 1.1
print(f"x₀ = {x}")

#  создание пустого списка для вывода
table_data = []
headers = ["n", "x_n", "|x_n - x_n-1|"]

#  вычисление значения
for n in range(1, 20):
    x_new = phi(x)
    diff = abs(x_new - x)
    
    table_data.append([n, f"{x_new:.8f}", f"{diff:.8f}"])
    
    #  вывод
    if diff < 1e-5:
        print(tabulate(table_data, headers=headers, tablefmt="plain", numalign="right"))
        print(f"Корень: x = {x_new:.8f}")
        print(f"Проверка: {x_new:.8f}·sin({x_new:.8f}) - 1 = {x_new * np.sin(x_new) - 1:.2e}")
        break
    x = x_new
```
