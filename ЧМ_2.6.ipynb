```python
#  определение функций
def f(x):
    return 4*x - np.log(x) - 4 if x > 0 else np.nan
def phi(x):
    return (4 + np.log(x)) / 4

#  решение методом простых итераций
x = 2.0
while True:
    x_new = phi(x)
    if abs(x_new - x) < 1e-4:
        break
    x = x_new

#  визуализация
x_vals = np.linspace(0.1, 3, 300)
plt.plot(x_vals, [f(v) for v in x_vals], 'b-')
plt.axhline(0, c='r', ls='--')
plt.plot(x_new, 0, 'ro', label=f'x = {x_new:.4f}')
plt.grid(alpha=0.3)
plt.title('4x - ln x = 4')
plt.legend()
plt.show()

#  вывод
print(f'Корень: {x_new:.4f}')
```
